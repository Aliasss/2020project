```python
import re
def cleaning(text):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
    result = hangul.sub('', text)
    return result

df['ko_text'] = df['tweet_text'].apply(lambda x: cleaning(x))
df.head()
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
korean_stopwords_path = 'C:/Users/seob6/Desktop/python&analysis/data_analysis_process/exercise/korean_stopwords.txt'
with open(korean_stopwords_path, encoding='utf-8') as f:
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]

def get_nouns(x):
    nouns_tagger = Okt()
    nouns = nouns_tagger.nouns(x)
    
    # 한 글자 키워드 제거
    nouns = [noun for noun in nouns if len(noun) > 1]
    # 불용어 제거
    nouns = [noun for noun in nouns if noun not in stopwords]
    return nouns

df['nouns'] = df['ko_text'].apply(lambda x: get_nouns(x))    
print(df.shape)
df.head()
```

```python
transactions = df['nouns'].tolist()
transactions = [transaction for transaction in transactions if transaction]
```

```python
print(transactions)
```

```python
results = list(apriori(transactions,
                      min_support=0.05,
                      min_confidence=0.2,
                      min_lift=1,
                      max_length=2))
print(results)
```

```python
network_df = pd.DataFrame(columns=['source', 'target', 'support'])

for result in results:
    if len(result.items) == 2:
        items = [x for x in result.items]
        row = [items[0], items[1], result.support]
        series = pd.Series(data=row, index=network_df.columns)
        network_df = network_df.append(series, ignore_index=True)
        
network_df.head()
```

```python
network_df.sort_values(by='support', ascending=False, inplace=True)
```

```python
tweet_corpus = "".join(df['ko_text'].tolist())

nouns_tagger = Okt()
nouns = nouns_tagger.nouns(tweet_corpus)
count = Counter(nouns)

remove_one_counter = Counter({x : count[x] for x in count if len(x) > 1})

node_df = pd.DataFrame(remove_one_counter.items(), columns=['node', 'node_size'])
node_df = node_df[node_df['node_size'] >= 50]
print(node_df.shape)
node_df.sort_values(by='node_size', ascending=False, inplace=True)
node_df
```

## 시각화

```python
import networkx as nx
plt.figure(figsize=(25, 25))

G = nx.Graph()
for index, row in node_df.iterrows():
    G.add_node(row['node'], nodesize=row['node_size'])
for index, row in network_df.iterrows():
    G.add_weighted_edges_from([(row['source'], row['target'], row['support'])])
    
pos = nx.spring_layout(G, k=0.6, iterations=50)
sizes = [G.node[node]['nodesize']*25 for node in G]
nx.draw(G, pos=pos, node_size=sizes)

nx.draw_networkx_labels(G, pos=pos, font_family='Malgun Gothic', font_size=25)

ax = plt.gca()
```
